* 熵：是一个物理学概念，表示一个系统的不确定性程度或者系统的混乱程度
* 信息熵：是将**熵**引入信息论中，用来衡量信息的不确定性
* 可以通过方程表示：$H(x) = -\sum_{i=1}^nP(i)logP(i)$
* 交叉熵：“如果把熵$H(x)$想象为<知道真实概率的人所经历的惊异程度>，那么什么是交叉熵？ 交叉熵从$P$到$Q$，记为$H(P,Q)$。我们可以把交叉熵想象为<主观概率为的观察者在看到根据概率生成的数据时的预期惊异>。”——————主要应用于度量同一随机变量$X$的预测分布和真实分布之间的差距，当两个分布相等时，交叉熵有最小值。
